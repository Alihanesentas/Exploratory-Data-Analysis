BOXPLOT


```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df["sex"].value_counts()
```

```python
df["smoker"].value_counts()
```

```python
df["day"].value_counts()
```

```python
df["time"].value_counts()
```

```python
#boxplot
sns.boxplot(x = df["total_bill"]);
```

```python
sns.boxplot(y = df["total_bill"], orient="v");
```

```python
#Which day's total bill is higher?
sns.boxplot(x = "day", y = "total_bill", data = df);
```

```python
#  dinner and lunch total bill comparison
sns.boxplot(x = "time", y = "total_bill", data = df);
```

```python
#  count of member and total bill comparison
sns.boxplot(x ="size",y="total_bill",data=df)
```

```python
# day and total bill and sex comparison
sns.boxplot(x="day",y="total_bill",hue="sex",data=df)
```

#Violinplot

```python
#Violinplot
df.head()
```

```python
sns.catplot(y="total_bill", kind="violin", data=df);
```

```python
#Violinplot crossover
sns.catplot(x="day", y="total_bill", kind="violin", data=df);
```

Scatter plot


```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", style = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", size = "size", data = df);
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
df = tips.copy()

```

```python
sns.lmplot(x = "total_bill", y = "tip", data = df,hue="smoker");
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = df,hue="smoker",col="time");
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = df,hue="smoker",col="time",row="sex");
```
